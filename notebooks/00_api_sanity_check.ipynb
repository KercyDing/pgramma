# 00 - API Connection Sanity Check

Verify that the OpenAI-compatible endpoint is reachable and responds correctly before building evaluation pipelines.

```python
from dotenv import load_dotenv
import os
from openai import OpenAI

load_dotenv("../.env")

client = OpenAI(
    base_url=os.environ["LLM_API_URL"],
    api_key=os.environ["LLM_API_KEY"],
)
```

```python
# List available models
models = client.models.list()
for m in models.data:
    print(m.id)
```

```python
MODEL = os.environ.get("LLM_MODEL", "gpt-5.1")

# Streaming chat completion test
stream = client.chat.completions.create(
    model=MODEL,
    messages=[{"role": "user", "content": "Say hello in one sentence."}],
    max_tokens=64,
    stream=True,
)
for chunk in stream:
    delta = chunk.choices[0].delta.content
    if delta:
        print(delta, end="", flush=True)
print()
```
